```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import itertools 
import numpy as np
from rkd.didactic import *
%matplotlib inline
```

```python
L1 = 50
L2 = 100
rr = Robot((L1,0,0,t1), (L2,0,0,t2))
```

```python
T = rr.T
xx = T[0,3]
yy = T[1,3]
```

```python
x = lambdify((t1,t2),xx)
y = lambdify((t1,t2),yy)
```

```python
_t1,_t2 = np.linspace(0, np.pi), np.linspace(-np.pi/2, np.pi/2)
points = itertools.product(_t1,_t2)
points = list( zip(*list(points) ))
T1,T2 = np.array(points[0]), np.array(points[1])
plt.plot(x(T1,T2), y(T1,T2), "b.")
plt.axis('square');
```
