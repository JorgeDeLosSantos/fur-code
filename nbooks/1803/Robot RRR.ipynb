```python
from rkd.didactic import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
d1 = 100
l2 = 150
l3 = 150
rb = Robot( (0,pi/2,d1,t1), (l2,0,0,t2), (l3,0,0,t3) )
```

```python
T = rb.T
```

```python
T10 = rb.Ts[0]
T20 = T10*rb.Ts[1]
O1 = T10[:3,3]
O2 = T20[:3,3]
```

```python
ecx = T[0,3]
ecy = T[1,3]
ecz = T[2,3]
```

```python
%matplotlib nbagg
import numpy as np
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

T2 = 0

for T1 in np.linspace(0, 2*np.pi, 10):
    for T3 in np.linspace(0, np.pi, 10):
        X = ecx.subs({t1:T1, t2:T2, t3:T3})
        Y = ecy.subs({t1:T1, t2:T2, t3:T3})
        Z = ecz.subs({t1:T1, t2:T2, t3:T3})
        O1N = O1.subs({t1:T1, t2:T2, t3:T3})
        O2N = O2.subs({t1:T1, t2:T2, t3:T3})
#         ax.plot([0, O1N[0], O2N[0], X], [0, O1N[1], O2N[1], Y], [0, O1N[2], O2N[2], Z], "r--")
        ax.plot([X], [Y], [Z], ".")
```
