```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import itertools 
import numpy as np
from rkd.didactic import *
%matplotlib nbagg
```

```python
d1 = 100
L2 = 200
L3 = 100
rrr = Robot((0,pi/2,d1,t1), (L2,0,0,t2), (L3,0,0,t3))
```

```python
T = rrr.T
xx = T[0,3]
yy = T[1,3]
zz = T[2,3]
```

```python
x = lambdify((t1,t2,t3),xx)
y = lambdify((t1,t2,t3),yy)
z = lambdify((t1,t2,t3),zz)
```

```python
_t1,_t2,_t3 = np.linspace(0, np.pi, 180), np.linspace(0, np.pi/2), np.linspace(-np.pi/2, np.pi/2)
points = itertools.product(_t1,_t2,_t3)
points = list( zip(*list(points) ))
T1,T2,T3 = np.array(points[0]), np.array(points[1]), np.array(points[2])
# plt.axis('square');
```

```python
figp = plt.figure()
axp = figp.add_subplot(111)
axp.plot(x(T1,T2,T3), y(T1,T2,T3), ".");
plt.axis('square');
```
