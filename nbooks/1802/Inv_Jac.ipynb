$$ J^+ = (J^T J)^{-1} J^T $$

```python
from furlib import *
```

```python
rr = Robot((l1,0,0,t1), (l2,0,0,t2))
rrr = Robot((0,pi/2,d1,t1), (0,pi/2,0,t2), (0,0,d3,0,'p'))
```

```python
vals = {l1:200, t1:pi/6, l2:200, t2:pi/6}
J = rr.J.subs(vals).evalf()
eta = Matrix([100,0,0,0,0,0])
Jpi = ( (J.T*J).inv() )* J.T
qp = Jpi*eta
qp, J*qp
```

```python
vals = {d1:100, t1:pi/4, t2:-pi/3, d3:110}
J = rrr.J.subs(vals).evalf()
eta = Matrix([200,-100,400,0,0,0])
Jpi = ( (J.T*J).inv() )* J.T
qp = Jpi*eta
qp, J*qp
```

## Para una cierta función

```python
# 
vx,vy = [],[]
J = rr.J
T = np.linspace(0,2*np.pi,100)
for t in T:
    JN = J.subs({l1:200, l2:200, t1:np.cos(t), t2:np.sin(t)})
    qp = Matrix([-np.sin(t), np.cos(t)])
    V = JN*qp
    vx.append(V[0])
    vy.append(V[1])
    
plt.plot(T, vx)
plt.plot(T, vy)
```
