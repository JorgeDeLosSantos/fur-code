```python
import numpy as np 
import matplotlib.pyplot as plt
from numpy import sin,cos,tan,arcsin,arccos,arctan,arctan2,sqrt,pi,exp
from mpl_toolkits.mplot3d import Axes3D
%matplotlib nbagg

def dh(a,alpha,d,theta):
    M = np.array([[cos(theta),-sin(theta)*cos(alpha),sin(theta)*sin(alpha),a*cos(theta)],
                  [sin(theta),cos(theta)*cos(alpha),-cos(theta)*sin(alpha),a*sin(theta)],
                  [0,sin(alpha),cos(alpha),d],
                  [0,0,0,1]])
    return M

def draw_uvw(H, ax):
    u = H[:3,0]
    v = H[:3,1]
    w = H[:3,2]
    o = H[:3,3]
    ax.quiver(o[0],o[1],o[2],u[0],u[1],u[2],color="r")
    ax.quiver(o[0],o[1],o[2],v[0],v[1],v[2],color="g")
    ax.quiver(o[0],o[1],o[2],w[0],w[1],w[2],color="b")
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
l1, l2 = 2, 2
t1, t2 = pi/6, pi/6

T1_0 = dh(l1,0,0,t1)
T2_1 = dh(l2,0,0,t2)
T2_0 = np.dot(T1_0, T2_1)
A = np.array([0,0,0])
B = T1_0[:3,3]
C = T2_0[:3,3]
ax.plot([A[0],B[0],C[0]], [A[1],B[1],C[1]], [A[2],B[2],C[2]], '-o')
draw_uvw(np.eye(4), ax)
draw_uvw(T1_0, ax)
draw_uvw(T2_0, ax)
ax.set_zlim([-2,2])
```
