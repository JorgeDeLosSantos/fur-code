```python
from rkd.didactic import *
import numpy as np
```

```python
r = Robot((0,pi/2,0,t1), (0,0,d2+100,0,"p"))
T = r.T
J = r.J
```

```python
t = np.linspace(0, 2*np.pi)
T1 = np.cos(t)
D2 = 10*t
Q1P = np.diff(T1)/np.diff(t)
Q2P = np.diff(D2)/np.diff(t)
```

```python
V = []
for k,_ in enumerate(t):
    if (k == len(t)-1): continue
    qp = Matrix([Q1P[k], Q2P[k]])
    JN = J.subs({t1:T1[k+1], d2:D2[k+1]}).evalf()
    V.append(JN*qp)
```

```python
plt.plot(t[1:],V)
```

```python
s = list( zip(*V) )
```

```python
plt.plot(t[1:], s[5])
```
